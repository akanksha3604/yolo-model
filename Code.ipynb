```python
# You can change to yolov8s.pt, yolov8m.pt, etc. for bigger models
model = YOLO("yolov8n.pt")
model.train(
    data="/content/data.yaml",  # <-- update this path
    epochs=50,       # maximum number of epochs
    imgsz=640,       # image size
    batch=16,        # batch size
    patience=5,      # early stopping (stop if no improvement for 5 epochs)
    save=True,       # save best & last model
    project="/content/drive/MyDrive/yolov8_runs",  # results folder in Drive
    name="exp1"      # experiment name
)
```
